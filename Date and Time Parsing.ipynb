```python
vl["Year"]=pd.to_datetime(vl["Last Known Eruption"], format="%y")
```

```python
volcanos = pd.DataFrame(vl["Last Known Eruption"])
def parse_eruption_year(eruption):
    if eruption == "Unknown":
        return None  # Handle unknown values as NaN (None)
    
    parts = eruption.split()
    year = int(parts[0])  # Extract year
    
    if len(parts) > 1 and parts[1] == "BCE":  # Convert BCE years to negative
        return -year
    return year  # CE years remain positive

vl["Parsed Eruption Year"] = vl["Last Known Eruption"].apply(parse_eruption_year)
```
